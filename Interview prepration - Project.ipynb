# Model Selection :

**There are several classification models are available, each with its strengths and weaknesses. I would like to choose Logistic regression, LDA (Linear Discremenant Analysis), KNN, Decision Tree, Random forest, Gradient boosting & XG Boost.**

Logistic Regression is often a good starting point due to its simplicity and interpretability, making it a strong choice for binary classification tasks. 

Decision Trees and Random Forests are excellent for handling complex, non-linear relationships within the data and can be more interpretable than deep learning models. 

For large-scale datasets or image classification, Gradient Boosting models, such as XGBoost, are known for their high accuracy and are popular choices in competitions.

## Logistic Regression

```python
# Importing Logistic Regression from Sklearn

from sklearn.linear_model import LogisticRegression
```

```python
# Creating an object for our model

clf_LR = LogisticRegression()
```

```python
# Fitting our model over in Train data set

clf_LR.fit(X_train, Y_train)
```

```python
# Predicitng value using our model over on test data.

Y_pred_LR = clf_LR.predict(X_test)
```

```python
# Calculating Accuracy score, Precision score, Recall score and printing them.

print("Accuracy score, logistic regression:",accuracy_score(Y_test, Y_pred_LR))
print("Precision score, logistic regression:", precision_score(Y_test, Y_pred_LR))
print("Recall score, logistic regression:", recall_score(Y_test, Y_pred_LR))
```

```python
# Calulating and displaying confusion matrix

confusion_matrix(Y_test, Y_pred_LR)
```

```python
# Plotting confusion matrix

df_cm = pd.DataFrame(confusion_matrix(Y_test, Y_pred_LR), range(2), range(2))
sns.set(font_scale= 1.2) # for label size
sns.heatmap(df_cm, annot=True, annot_kws={"size": 14}) # font size

plt.show()
```

```python
# Plotting ROC curve

fpr, tpr, _ = metrics.roc_curve(Y_test,  Y_pred_LR)
auc = metrics.roc_auc_score(Y_test, Y_pred_LR) # Calculating and plotting auc value
plt.plot(fpr,tpr,label="LR, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

### Model Interpretation :

**Accuracy Score (0.8248):** Our logistic regression model has an accuracy of approximately 82.48%. This means that it correctly predicts about 82.48% of the instances in our dataset. While accuracy is a good overall measure of model performance, it should be considered in conjunction with other metrics for a more complete assessment.

**Precision Score (0.7445):** The precision score of 0.7445 indicates that when our model predicts a positive outcome, it is correct about 74.45% of the time. In other words, the model's positive predictions are relatively accurate. High precision is important when false positives are costly or undesirable.

**Recall Score (0.5543):** The recall score of 0.5543 tells us that our model correctly identifies approximately 55.43% of all actual positive instances. This is particularly important when missing positive cases (false negatives) is a concern.

**ROC AUC Score (0.7410):** The ROC AUC score of 0.7410 is a measure of the model's ability to distinguish between the positive and negative classes. A value of 0.7410 indicates that our model performs reasonably well in this regard.

### Model Interpretation :



when using grid search to optimize the number of neighbors (increasing from 1 to 10), the KNN model shows improved accuracy and precision but a decreased recall. The choice of the best value for n_neighbors should consider the specific goals and trade-offs of our classification task. Depending on your objectives, We may prioritize precision, recall, or a balanced approach.

**To determine the best model, we should consider our specific use case:**

If our task prioritizes accurately identifying positive cases (high recall) and we are willing to accept some false positives, then the KNN model with n_neighbors=1 might be preferred.

If our task requires a higher degree of overall accuracy and precision and we can tolerate a lower recall, then the Grid Search KNN model with n_neighbors=10 might be the better choice.
